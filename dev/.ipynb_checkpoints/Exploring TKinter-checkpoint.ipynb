# Exploring TKinter for Bit

## TK Basics

```python
import tkinter as tk
from tkinter import ttk
import matplotlib.pyplot as plt
```

```python
root = tk.Tk()
root.title('Testing')
root.mainloop()
```

You can run more than one mainloop in a single python process, unlike PyQt. 

This would let us include more than one `@Bit.run` function per file (not that we need this feature, but it is allowed now).

```python
# Thank you: https://www.geeksforgeeks.org/creating-tabbed-widget-with-python-tkinter/

root = tk.Tk()
root.title("Tab Widget")
tabControl = ttk.Notebook(root)
  
tab1 = ttk.Frame(tabControl)
tab2 = ttk.Frame(tabControl)
  
tabControl.add(tab1, text ='Tab 1')
tabControl.add(tab2, text ='Tab 2')
tabControl.pack(expand = 1, fill ="both")
  
ttk.Label(tab1, 
          text ="stuff").grid(column = 0, 
                               row = 0,
                               padx = 30,
                               pady = 30)  
ttk.Label(tab2,
          text ="more stuff").grid(column = 0,
                                    row = 0, 
                                    padx = 30,
                                    pady = 30)
  
root.mainloop() 
```

## Integration with Matplotlib

```python
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg

from matplotlib.figure import Figure
import numpy as np
```

```python
root = tk.Tk()
root.title("Embedding in Tk")

fig = Figure(figsize=(5, 4), dpi=100)
t = np.arange(0, 3, .01)
ax = fig.add_subplot()
line, = ax.plot(t, 2 * np.sin(2 * np.pi * t))
ax.set_xlabel("time [s]")
ax.set_ylabel("f(t)")
fig.tight_layout(pad=0.5)

canvas = FigureCanvasTkAgg(fig, master=root)  # A tk.DrawingArea.
canvas.draw()

def update_frequency(new_val):
    # retrieve frequency
    f = float(new_val)

    # update data
    y = 2 * np.sin(2 * np.pi * f * t)
    line.set_data(t, y)

    # required to update canvas and attached toolbar!
    canvas.draw()


slider_update = tk.Scale(root, from_=1, to=5, orient=tk.HORIZONTAL,
                              command=update_frequency, label="Frequency [Hz]")

# Packing order is important. Widgets are processed sequentially and if there
# is no space left, because the window is too small, they are not displayed.
# The canvas is rather flexible in its size, so we pack it last which makes
# sure the UI controls are displayed as long as possible.
slider_update.pack(side=tk.BOTTOM)
canvas.get_tk_widget().pack(side=tk.TOP, fill=tk.BOTH, expand=1)

tk.mainloop()
```

It looks like embedding the plot should be straightforward.

## Controls + canvas

```python
class CounterCanvas(FigureCanvasTkAgg):
    def __init__(self, root):
        self.counter = 0
        self.figure = Figure(figsize=(5,4), dpi=100)
        self.axes = self.figure.add_subplot()
        super(FigureCanvasTkAgg, self).__init__(self.figure, master=root)

    def refresh(self):
        self.axes.clear()
        self.axes.bar(1, self.counter)
        self.axes.set_ylim([0, 10])
        self.axes.set_ylabel("Count")
        self.draw()
        
    def increment(self):
        self.counter += 1
        self.refresh()
        
    def decrement(self):
        self.counter -= 1
        self.refresh()

        
root = tk.Tk()
root.title('Counter')

def add_counter_canvas(master):
    canvas = CounterCanvas(master)

    button_frame = tk.Frame(master, width=500, height=100)
    increment_button = tk.Button(master=button_frame, text="Increment", command=canvas.increment)
    decrement_button = tk.Button(master=button_frame, text="Decrement", command=canvas.decrement)
    increment_button.pack(side=tk.LEFT)
    decrement_button.pack(side=tk.RIGHT)

    button_frame.pack(side=tk.BOTTOM)
    canvas.get_tk_widget().pack(side=tk.TOP, fill=tk.BOTH, expand=1)

add_counter_canvas(root)
tk.mainloop()
```

## Multiple Tabs

```python
class CounterCanvas(FigureCanvasTkAgg):
    def __init__(self, root):
        self.counter = 0
        self.figure = Figure(figsize=(5,4), dpi=100)
        self.axes = self.figure.add_subplot()
        super(FigureCanvasTkAgg, self).__init__(self.figure, master=root)
        self.draw()

    def refresh(self):
        self.axes.clear()
        self.axes.bar(1, self.counter)
        self.axes.set_ylim([0, 10])
        self.axes.set_ylabel("Count")
        self.draw()
        
    def increment(self):
        self.counter += 1
        self.refresh()
        
    def decrement(self):
        self.counter -= 1
        self.refresh()

        
def add_counter_canvas(master):
    canvas = CounterCanvas(master)

    button_frame = tk.Frame(master, width=500, height=100)
    increment_button = tk.Button(master=button_frame, text="Increment", command=canvas.increment)
    decrement_button = tk.Button(master=button_frame, text="Decrement", command=canvas.decrement)
    increment_button.pack(side=tk.LEFT)
    decrement_button.pack(side=tk.RIGHT)

    button_frame.pack(side=tk.BOTTOM)
    canvas.get_tk_widget().pack(side=tk.TOP, fill=tk.BOTH, expand=1)

    
root = tk.Tk()
root.title('Counters')

style = ttk.Style(root)
style.configure('TNotebook', tabposition='s')

tabControl = ttk.Notebook(root, width=500, height=500, style='TNotebook')
  
tab1 = ttk.Frame(tabControl)
tab2 = ttk.Frame(tabControl)
  
tabControl.add(tab1, text ='Counter 1')
tabControl.add(tab2, text ='Counter 2')
tabControl.pack(expand = 1, fill ="both")

add_counter_canvas(tab1)
add_counter_canvas(tab2)

tk.mainloop()
```
